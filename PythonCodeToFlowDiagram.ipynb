````python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
````

````python
# Step 1: Load the pre-trained CodeT5 model and tokenizer
model_name = "Salesforce/codet5-base"  # You can try other variants like 'codet5-large' for more complexity
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
````

````python
# Step 2: Define the Python code input
python_code = """
def example(x):
    if x > 0:
        print("Positive")
    else:
        print("Non-positive")
"""
````

````python
# Step 3: Tokenize the input code
inputs = tokenizer(
    python_code,
    return_tensors="pt",  # Convert input into PyTorch tensors
    max_length=512,       # Ensure input does not exceed model limit
    truncation=True       # Truncate if input is too long
)
````

````python
# Step 4: Generate the output
outputs = model.generate(
    inputs["input_ids"],
    max_length=100,       # Limit the output length
    temperature=0.7,      # Control creativity of output
    top_p=0.9,            # Use nucleus sampling
    num_beams=5           # Beam search for better results
)
````

````python
# Step 5: Decode the output to human-readable text
summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
````

````python
# Step 6: Print the result
print("Generated Summary of Code Logic:")
print(summary)
````

````python
from transformers import RobertaTokenizer, RobertaModel

# Load the model and tokenizer
tokenizer = RobertaTokenizer.from_pretrained("microsoft/graphcodebert-base")
model = RobertaModel.from_pretrained("microsoft/graphcodebert-base")

# Example Python code
code = """
def example(x):
    if x > 0:
        return "Positive"
    else:
        return "Negative"
"""

# Tokenize input code
inputs = tokenizer(code, return_tensors="pt", max_length=512, truncation=True)

# Generate embeddings
outputs = model(**inputs)
embeddings = outputs.last_hidden_state

````

````python
print("Code embeddings shape:", embeddings.shape)
````

````python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("TroyDoesAI/MermaidStable3B")
model = AutoModelForCausalLM.from_pretrained("TroyDoesAI/MermaidStable3B")
````

````python
code_input = """
def example(x):
    if x > 0:
        print("Positive")
    elif x == 0:
        print("Zero")
    else:
        print("Negative")
    return x * 2
"""
````

````python
# Create the prompt with instruction
prompt = f"### Instruction: Create the mermaid diagram for the following code:\n{code_input}\n### Response:\n```mermaid\n"

# Tokenize the input
inputs = tokenizer(prompt, return_tensors="pt")

# Generate the output
outputs = model.generate(inputs["input_ids"], max_length=512, temperature=0.7)

# Decode the output
mermaid_diagram = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Append closing backticks for code block
mermaid_diagram += "\n```"
````
